### Visualisasi Outfull

```python
# konvert kolom time_ref pada data revised menjadi tipe string dan bisa dikonvert ke datetime dengan cepat
# dieksekusi sekali saja agar tidak terjadi penambahan secara berlebihan

data_outfull['time_ref'] = data_outfull['time_ref'].apply(lambda x:str(x)+'01')

#ubah type data kolom time_ref ke timestamp

data_outfull['time_ref'] = [pd.Timestamp(v) for i,v in data_outfull['time_ref'].items()]

data_outfull
```

```python
data_outfull.groupby('time_ref')['value'].max()
```

```python
import seaborn as sb

# visualisasi
py.style.use('seaborn')

fig = py.figure()
ax = fig.subplots()
ax.set_title('Data outfull \nberdasarkan seleksi max dan min \npada kolom time_ref\n',fontweight='bold',fontsize=14)

#ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.plot(data_outfull.groupby('time_ref')['value'].max())
ax.plot(data_outfull.groupby('time_ref')['value'].min())

# create df for value max and min
doutfull_max_time_ref = data_outfull.groupby('time_ref')['value'].max()
doutfull_min_time_ref = data_outfull.groupby('time_ref')['value'].min()

jangkauan = doutfull_max_time_ref - doutfull_min_time_ref

# max
#(x,bottom max),(y,top max)
ax.text(pd.Timestamp(doutfull_max_time_ref.sort_values(ascending=False).index[0]),max(doutfull_max_time_ref),'MAX\n'+str(max(doutfull_max_time_ref)),ha='center',va='top')
ax.scatter(pd.Timestamp(doutfull_max_time_ref.sort_values(ascending=False).index[0]),max(doutfull_max_time_ref))

# min
ax.text(pd.Timestamp(doutfull_min_time_ref.sort_values().index[0]),min(doutfull_min_time_ref),'MIN\n'+str(min(doutfull_min_time_ref)),ha='center',va='bottom')
ax.scatter(pd.Timestamp(doutfull_min_time_ref.sort_values().index[0]),min(doutfull_min_time_ref))

# jangkauan terjauh
xyA = (jangkauan.sort_values(ascending=False).index[0],
       doutfull_max_time_ref.loc[doutfull_max_time_ref.index == jangkauan.sort_values(ascending=False).index[0]][0])
xyB = (jangkauan.sort_values(ascending=False).index[0],
       doutfull_min_time_ref.loc[doutfull_min_time_ref.index == jangkauan.sort_values(ascending=False).index[0]][0])
linemax = patch.ConnectionPatch(xyA = xyA ,xyB= xyB,coordsA='data',linestyle='--')
ax.add_artist(linemax)

ax.text(jangkauan.sort_values(ascending=False).index[0],
       doutfull_max_time_ref.loc[doutfull_max_time_ref.index == jangkauan.sort_values(ascending=False).index[0]][0] / 3,'Jangkauan terjauh\n'+str(jangkauan.max()),ha='center')

# jangkauan terdekat
xyA = (jangkauan.sort_values().index[0],
       doutfull_max_time_ref.loc[doutfull_max_time_ref.index == jangkauan.sort_values().index[0]][0])
xyB = (jangkauan.sort_values().index[0],
       doutfull_min_time_ref.loc[doutfull_min_time_ref.index == jangkauan.sort_values().index[0]][0])
linemax = patch.ConnectionPatch(xyA = xyA ,xyB= xyB,coordsA='data',linestyle='--')
ax.add_artist(linemax)

ax.text(jangkauan.sort_values().index[0],
       doutfull_max_time_ref.loc[doutfull_max_time_ref.index == jangkauan.sort_values().index[0]][0] / 3,'Jangkauan terdekat\n'+str(jangkauan.min()),ha='center')

label = ax.get_xticklabels()
py.setp(label,rotation=45,ha='right')
ax.legend(['Max','Min'])

py.show()
```

```python
# set data
tgl = 2016
cektgl = data_outfull['time_ref'].apply(lambda x:x.year == tgl)

#print(data_revised[cektgl])
data_out_tgl = data_outfull[cektgl]

#visualisasi
fig, ax = py.subplots()
ax.set_title('Data Outfull pada tahun '+str(tgl),fontweight='bold',fontsize=14)

ax.plot(data_out_tgl['value'])

py.show()
```
